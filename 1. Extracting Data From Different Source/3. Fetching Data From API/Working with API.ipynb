```python
# at a time only 20 rows are there in 1 page.
# so we have to do this in the loop.
# There are 532 pages. , total_results = 10625 rows
```

```python
# First Create empty DataFrame
temp_df = pd.DataFrame()
```

```python
temp_df
```

```python
# now, we hit each time and append that our temp_df
```

```python
for i in range(1,51):
    response = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=2a97d277bcad521c86a18d6f78ff9a4e&language=en-US&page={}'.format(i))
    df = pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','popularity','vote_average','vote_count']]
    temp_df = pd.concat([temp_df,df],ignore_index = True)
```
