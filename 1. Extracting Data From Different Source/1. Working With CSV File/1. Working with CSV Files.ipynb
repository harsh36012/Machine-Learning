### Opening a csv file from an URL (Server)

```python
# this is the code 
```

```python
import requests

url = 'https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv' # Replace with your actual URL
local_filename = 'downloaded_data2.csv'

try:
    # Send a GET request to the URL
    response = requests.get(url)
    response.raise_for_status() # Check for bad responses (e.g., 404, 500)

    # Write the content to a local file in binary mode ('wb')
    with open(local_filename, 'wb') as f:
        f.write(response.content)
    print(f"File successfully saved to {local_filename}")

except requests.exceptions.RequestException as e:
    print(f"An error occurred: {e}")
```

### skip bad lines (on_bad_lines = 'skip')

```python
pd.read_csv('BX-Books.csv', sep=';' , encoding="latin-1")
```

```python
# ParserError: Error tokenizing data. C error: Expected 1 fields in line 6, saw 2
# this error occur when, some of the tokens can't be converted into another encoding format.
# so these are bad lines, so we have to skip those lines

pd.read_csv('BX-Books.csv', sep=',' , encoding="latin-1", on_bad_lines = 'skip')
```

### converters

```python
pd.read_csv('IPL Matches 2008-2020.csv')
```

```python
def rename(name):
    if name == 'Royal Challengers Bangalore':
        return 'RCB'
    else :
        return name
```

```python
rename('Royal Challengers Bangalore')
```

```python
pd.read_csv('IPL Matches 2008-2020.csv', converters= {'team1':rename})
```
